```python
p_data9 = pd.read_csv('../../handle_Data/book_kind/book_英语_2020-01-09.csv')
p_data9 = pd.DataFrame(p_data9)
p_data10 = pd.read_csv('../../handle_Data/book_kind/book_自我管理_2020-01-09.csv')
p_data10 = pd.DataFrame(p_data10)
p_data11 = pd.read_csv('../../handle_Data/book_kind/book_诗歌_2020-01-09.csv')
p_data11 = pd.DataFrame(p_data11)
p_data12 = pd.read_csv('../../handle_Data/book_kind/book_爱情_职场_2020-01-09.csv')
p_data12 = pd.DataFrame(p_data12)

data1 = pd.concat([data,p_data9,p_data10,p_data11,p_data12])
data1 = data1.reset_index()
data1 = data1.drop('index',axis=1)

mean_kind2 = data1.groupby('分类').count()
mean_kind2 = pd.DataFrame(mean_kind2['书名'])


mean_kind2
name = list(mean_kind2.index)
value1 = []
for i in mean_kind2.values:
    value1.append(int(i[0]))
words = []
for i in range(0,len(name)):
    words.append((name[i],value1[i]))
words
```

```python
from pyecharts import options as opts
from pyecharts.charts import Page, WordCloud
from pyecharts.globals import SymbolType


c = (
        WordCloud()
        .add("", words, word_size_range=[20, 100])
        .set_global_opts(title_opts=opts.TitleOpts(title="种类词云"))
    )
    
c.render_notebook()
```
